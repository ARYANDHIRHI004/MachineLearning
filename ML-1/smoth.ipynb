```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_clusters_per_class=1, weights=[0.90], n_redundant=0, random_state=1)
```

```python
import pandas as pd 
```

```python
df1 = pd.DataFrame(X, columns=['f1','f2'])
df2 = pd.DataFrame(y, columns=["target"])
```

```python
final = pd.concat([df1, df2], axis=1)
```

```python
final
```

```python
final["target"].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final['f1'], final['f2'], c=final['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## transform the dataset 
oversample = SMOTE()
X,y = oversample.fit_resample(final[['f1', 'f2']], final[['target']])
```
