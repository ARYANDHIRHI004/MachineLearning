## selecting and extracting the most importance features from raw data 

### 1.Standardization 

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
totalBill = df['total_bill']
```

```python
import numpy as np 
mean = np.mean(totalBill)
std = np.std(totalBill)
```

```python
normalizedData = []
for i in totalBill:
    z_score = (i-mean)/std
    normalizedData.append(z_score)

print(normalizedData)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```
